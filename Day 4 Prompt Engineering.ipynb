```python
prompt = """
Recipe: Classic Chocolate Chip Cookies

Ingredients:
- 1 cup unsalted butter, softened
- 1 cup granulated sugar
- 1 cup brown sugar, packed
- 2 large eggs
- 1 teaspoon vanilla extract
- 3 cups all-purpose flour
- 1 teaspoon baking soda
- 1/2 teaspoon salt
- 2 cups chocolate chips

Instructions:
1. Preheat the oven to 375°F (190°C).
2. In a large mixing bowl, cream together the softened butter, granulated sugar, and brown sugar until light and fluffy.
3. Beat in the eggs one at a time, followed by the vanilla extract.
4. In a separate bowl, whisk together the flour, baking soda, and salt.
5. Gradually add the dry ingredients to the butter mixture, mixing until well combined.
6. Stir in the chocolate chips.
7. Drop rounded tablespoonfuls of dough onto ungreased baking sheets.
8. Bake for 9-11 minutes or until golden brown around the edges.
9. Allow the cookies to cool on the baking sheets for a few minutes, then transfer to wire racks to cool completely.

Enjoy these delicious classic chocolate chip cookies!
"""

response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": prompt}],
    temperature=0,
)

print(response.choices[0].message["content"])
```

```python
prompt = """
Recipe: Creamy Garlic Parmesan Pasta

Ingredients:
- 8 oz spaghetti
- 2 tablespoons butter
- 4 cloves garlic, minced
- 1 cup heavy cream
- 1 cup grated Parmesan cheese
- Salt and pepper, to taste
- Fresh parsley, chopped (for garnish)

Instructions:
1. Cook the spaghetti according to the package instructions. Drain and set aside.
2. In a large skillet, melt the butter over medium heat.
3. Add the minced garlic and sauté until fragrant, about 1-2 minutes.
4. Pour in the heavy cream and bring to a simmer.
5. Stir in the grated Parmesan cheese until melted and well combined.
6. Season with salt and pepper to taste.
7. Add the cooked spaghetti to the sauce and toss until evenly coated.
8. Remove from heat and garnish with chopped parsley.
9. Serve hot and enjoy!

Indulge in this creamy and flavorful garlic Parmesan pasta!
"""

response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": prompt}],
    temperature=0,
)

print(response.choices[0].message["content"])
```

```python
prompt = """
Recipe: Homemade Pizza

Ingredients:
- 1 1/2 cups warm water
- 2 1/4 teaspoons active dry yeast
- 1 tablespoon honey
- 3 1/2 cups all-purpose flour
- 2 tablespoons olive oil
- 2 teaspoons salt
- 1 cup pizza sauce
- 2 cups shredded mozzarella cheese
- Toppings of your choice (e.g., pepperoni, mushrooms, bell peppers)

Instructions:
1. In a large mixing bowl, combine the warm water, yeast, and honey. Let it sit for 5 minutes or until foamy.
2. Add the flour, olive oil, and salt to the yeast mixture. Stir until a sticky dough forms.
3. Transfer the dough to a floured surface and knead for about 5 minutes until smooth and elastic.
4. Place the dough in a greased bowl, cover with a damp cloth, and let it rise for 1-2 hours or until doubled in size.
5. Preheat the oven to 475°F (245°C).
6. Punch down the dough and divide it in half. Roll out each portion into a desired pizza shape.
7. Transfer the pizza dough to a baking sheet or pizza stone.
8. Spread pizza sauce evenly over the dough, leaving a small border around the edges.
9. Sprinkle shredded mozzarella cheese over the sauce and add your favorite toppings.
10. Bake in the preheated oven for 12-15 minutes or until the crust is golden and the cheese is bubbly.
11. Remove from the oven, let it cool for a few minutes, then slice and serve.

Savor the deliciousness of homemade pizza with your preferred toppings!
"""

response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": prompt}],
    temperature=0,
)

print(response.choices[0].message["content"])
```

```python
# Define the menu options
menu_options = {
    "1": "Get weather forecast",
    "2": "Exit",
}

# Display the menu options
print("Menu Options:")
for key, value in menu_options.items():
    print(f"{key}. {value}")

# Prompt the user to select an option
user_input = input("Please select an option (1-2): ")

# Handle user input
if user_input == "1":
    # Generate a prompt for weather forecast
    prompt = "What location would you like to get the weather forecast for?"
    # Make an API call to retrieve the weather forecast
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7,
        max_tokens=50,
    )
    print(response.choices[0].message["content"])
elif user_input == "2":
    # Exit the menu
    print("Goodbye!")
else:
    # Invalid input
    print("Invalid option. Please try again.")
```

```python
# Define the menu options
menu_options = {
    "1": "Addition",
    "2": "Subtraction",
    "3": "Multiplication",
    "4": "Exit",
}

# Display the menu options
print("Menu Options:")
for key, value in menu_options.items():
    print(f"{key}. {value}")

# Prompt the user to select an option
user_input = input("Please select an option (1-4): ")

# Handle user input
if user_input in ["1", "2", "3"]:
    # Prompt the user for numbers
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))

    if user_input == "1":
        # Perform addition
        result = num1 + num2
        print("Result: ", result)
    elif user_input == "2":
        # Perform subtraction
        result = num1 - num2
        print("Result: ", result)
    elif user_input == "3":
        # Perform multiplication
        result = num1 * num2
        print("Result: ", result)
elif user_input == "4":
    # Exit the menu
    print("Goodbye!")
else:
    # Invalid input
    print("Invalid option. Please try again.")
```
